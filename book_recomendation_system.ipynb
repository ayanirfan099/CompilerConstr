```python
def recommend_books(genre_input, price_condition='Free', num_recommendations=5):


    genre_encoder = LabelEncoder()
    genre_encoder.fit(book['Genre'])
    genre_encoded = genre_encoder.transform([genre_input])[0]


    filtered_books = book[(book['Genre'] == genre_input)]


    if filtered_books.empty:
        return f"No books found for genre '{genre_input}' that are '{price_condition}'."


    recommendations = filtered_books[['Book', 'Author(s)','Genre']].head(num_recommendations)

    return recommendations
genre_to_search = (input("enter your taste"))  # Example: Looking for fantasy books
recommended_books = recommend_books(genre_to_search, price_condition='Free', num_recommendations=5)
print("Recommended Free Books in the 'Fantasy' Genre:\n", recommended_books)
```
